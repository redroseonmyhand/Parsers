```python
assets_modified = []
revenue_modified = []
p_e_modified = []
roa_modified = []
ev_ebitda_modified = []
d_e_modified = []
for i in assets:
    assets_modified.append(int(str(i).replace(",", "")))
for i in revenue:
    revenue_modified.append(int(str(i).replace(",", "")))
for i in p_e:
    try:
        p_e_modified.append(float(i))
    except:
        p_e_modified.append(i)
for i in roa:
    try:
        roa_modified.append(float(i.replace('%', '')) / 100)
    except:
        roa_modified.append(i)
for i in ev_ebitda:
    try:
        ev_ebitda_modified.append(float(i))
    except:
        ev_ebitda_modified.append(i)
for i in d_e:
    try:
        d_e_modified.append(float(i.replace('%', '').replace(',', '')) / 100)
    except:
        d_e_modified.append(i)
df["total_assets"] = assets_modified
df["total_revenue"] = revenue_modified
df["p_e"] = p_e_modified
df["roa"] = roa_modified
df["ev_ebitda"] = ev_ebitda_modified
df["d_e"] = d_e_modified
```
